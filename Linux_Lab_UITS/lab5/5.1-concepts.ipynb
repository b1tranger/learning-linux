`if-elif-else` Syntax 

For multiple conditions, you can use `elif` (else if) statements. 

bash

    if [ condition1 ]; then
      # Commands if condition1 is true
    elif [ condition2 ]; then
      # Commands if condition1 is false and condition2 is true
    else
      # Commands if all conditions are false
    fi

<hr>
The standard syntax for a Bash `for` loop is: 

bash

    for variable_name in [list of items]; do
      # Commands to be executed for each item
    done

<hr>
**iterating over a range of numbers:**

bash

    #!/bin/bash
    for i in {1..5}; do
      echo "Number $i"
    done

<hr>
The basic syntax of a `while` loop is: 

bash

    while [ condition ]
    do
      # Commands to execute
    done
