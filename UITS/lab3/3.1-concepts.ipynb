## # how do I run a python file (test.py) using test.sh




- To run a Python file (`test.py`) using a shell script (`test.sh`), you need to create the shell script and make it executable.
- First, ensure you have your Python script ready: 
### [file](Lab_Tasks/Lab_3/1/print-hello.py)
```py
print("Hello World")
```
- Next, create the test.sh file and add the command to execute your Python script: 
  ### [file](Lab_Tasks/Lab_3/1/run-py.sh)
  
```sh
python3 test.py
```
- The `#!/bin/bash` line (known as a shebang) tells the system to run this file using the bash interpreter.
- `python3 test.py` is the command that executes your Python file. (You might need to use `python` instead of `python3` depending on your system's configuration).
- You should now be able to run the script successfully
```sh
./test.sh
```
- The `zsh: permission denied: ./test.sh` error indicates that your user account lacks the necessary execute permissions to run the test.sh file. This is a common issue when creating new scripts. 
- use the `ls -l` command to view the file's current permissions: 
```sh
ls -l test.sh
```
The output will look something like `-rw-r--r--`. The third character group relates to the file owner's permissions. The absence of an x (execute) in that group (e.g., `-rw-r--r--`) means it cannot be executed as a program. 
- Use the chmod command to add execute permission to the file. 
```sh
chmod u+x test.sh
```
To grant execute permission to all users
```sh
chmod +x test.sh
```
- alternative execution commands
```sh
bash test.sh
```

In Bash, input and output are managed via standard streams (stdin, stdout, stderr) and special commands like read, echo, and printf, along with powerful redirection operators and pipes. 
- Interactive Input (read command): The read command pauses the script and waits for user input from the keyboard, storing it in a variable.
- example:
```sh
echo "Enter your name:"
read name
echo "Hello, $name!"
```
- With a prompt: The -p option displays a prompt before waiting for input on the same line.
```sh
read -p "Enter your age: " age
echo "You are $age years old."
```
### example: [file](Lab_Tasks/Lab_3/2/read-echo.sh)

- Silent input: The -s option hides input for sensitive data like passwords
```sh
read -sp "Password: " password
echo # prints a newline after the hidden input
```
